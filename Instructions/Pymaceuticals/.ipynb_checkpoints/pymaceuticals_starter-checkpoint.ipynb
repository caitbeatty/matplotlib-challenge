## Observations and Insights 

Reviewing the following data, observations can be made that Capomulin and Ramicane are initially quite close in many respects in terms of statistical values such as mean and variance. This tells us these drugs perform consistent across samples. The population was well represented in the mice across all drug regimens. Ramicane would be a competitor to Capomulin. We also see a correlation between weight and tumor volume size. This should be studied more to understand if the larger the weight, the less impact Capomulin can make. Further insights should continue to be explored. 

```python
# Dependencies and Setup
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)


# Combine the data into a single dataset
whole_df = pd.merge(mouse_metadata, study_results, how = "outer", on="Mouse ID")


# Display the data table for preview
whole_df.head(10)
```

```python
#mouse count
count= len(whole_df["Mouse ID"].unique())
count
```

```python
#Find duplicate mouse
duplicatemouse = whole_df[whole_df.duplicated(['Mouse ID', 'Timepoint'], keep = False)]
duplicatemouse = duplicatemouse.set_index('Mouse ID')
duplicatemouse
```

```python
# drop duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
clean_df= whole_df.set_index('Mouse ID')
clean_df= clean_df. drop(duplicatemouse.index)
clean_df.reset_index(inplace=True)
clean_df
#whole_df.drop_duplicates()
```

```python
# Checking the number of mice in the clean DataFrame.
count = len(clean_df["Mouse ID"].unique())
count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

## Use groupby and summary statistical methods to calculate the following properties of each drug regimen: mean  median, variance, standard deviation, and SEM of the tumor volume



drug_group = clean_df.groupby("Drug Regimen")
drug_group_mean = drug_group[["Tumor Volume (mm3)"]].mean()
#drug_group_mean

                            
```

```python
#median, 
drug_group_median = drug_group[["Tumor Volume (mm3)"]].median()
#drug_group_median
```

```python
#variance, 
drug_group_variance = drug_group[["Tumor Volume (mm3)"]].var()
#drug_group_variance
```

```python
#standard deviation, and 

drug_group_standard_dev = drug_group[["Tumor Volume (mm3)"]].std()
#drug_group_standard_dev
```

```python
#SEM of the tumor volume. 
drug_group_SEM = drug_group[["Tumor Volume (mm3)"]].sem()
#drug_group_SEM

```

```python
# Assemble the resulting series into a single summary dataframe.
drug_summary_2_df = pd.merge(drug_group_median, drug_group_variance, how = "inner", on="Drug Regimen")
drug_summary_2_df= drug_summary_2_df.rename(columns={"Tumor Volume (mm3)_x" : "Mean Tumor Volume (mm3)", "Tumor Volume (mm3)_y":"Median Tumor Volume (mm3)"})

drug_summary_3_df= pd.merge(drug_summary_2_df, drug_group_variance, how = "inner", on="Drug Regimen" )
drug_summary_3_df= drug_summary_3_df.rename(columns={"Tumor Volume (mm3)" : "Variance of Tumor Volume (mm3)"})


drug_summary_4_df= pd.merge(drug_summary_3_df, drug_group_standard_dev, how = "inner", on="Drug Regimen" )
drug_summary_4_df= drug_summary_4_df.rename(columns={"Tumor Volume (mm3)" : "Standard Dev of Tumor Volume (mm3)"})

drug_summary_all_df= pd.merge(drug_summary_4_df, drug_group_SEM, how = "inner", on="Drug Regimen" )
drug_summary_all_df

drug_summary_all_df= drug_summary_all_df.rename(columns={"Tumor Volume (mm3)" : "SEM Tumor Volume (mm3)"})
drug_summary_all_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
drug_summary_short_df= drug_group["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem' ])


drug_summary_short_df= drug_summary_short_df.rename(columns={"mean": "Mean Tumor Volume (mm3)", "median":"Median Tumor Volume (mm3)", "var":"Variance of Tumor Volume (mm3)", "std":"Standard Dev of Tumor Volume (mm3)", "sem":"SEM Tumor Volume (mm3)"})
drug_summary_short_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  

# Capomulin, Ramicane, Infubinol, and Ceftamin

tumor_volume_df= clean_df[["Drug Regimen", "Timepoint", "Mouse ID"]]

# Start by getting the last (greatest) timepoint for each mouse
group_drug= tumor_volume_df.groupby(["Drug Regimen", "Mouse ID"])
maximums_df= group_drug.max()
maximums_df= maximums_df.reset_index()
maximums_df


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
mouse_final_time = pd.merge(clean_df, maximums_df, how = "inner", on=["Mouse ID", "Timepoint"])
mouse_final_time
mouse_final_time= mouse_final_time.rename(columns={"Drug Regimen_x" : "Drug Regimen"})

filter_mouse_final= mouse_final_time[["Mouse ID", "Drug Regimen", "Timepoint", "Tumor Volume (mm3)"]]
filter_mouse_final
```

```python
# Put treatments into a list for for loop (and later for plot labels)

#regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
#volumes= []


#capomulin_df = filter_mouse_final[filter_mouse_final["Drug Regimen"] == "Capomulin"]
#capomulin_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)

regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
volumes= []


#capomulin_df = filter_mouse_final[filter_mouse_final["Drug Regimen"] == "Capomulin"]
#capomulin_df
# Locate the rows which contain mice on each drug and get the tumor volumes
for drug in regimens:
    tumor_volume = filter_mouse_final.loc[filter_mouse_final["Drug Regimen"]==drug]["Tumor Volume (mm3)"]
    volumes.append(tumor_volume)
    quartiles = tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
      
    #print(f"The lower quartile of tumor volume is: {lowerq}")
    #print(f"The upper quartile of tumor volumeis: {upperq}")
    #print(f"The interquartile range of tumor volume is: {iqr}")
    #print(f"The the median of tumor volume is: {quartiles[0.5]} ")
    
    #Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    #print(f"Values below {lower_bound} could be outliers.")
    #print(f"Values above {upper_bound} could be outliers.")
    
    outlier_tumor_volume = tumor_volume.loc[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    print(drug, outlier_tumor_volume)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1= plt.subplots()
ax1.set_title("Tumor volume across regimens")
ax1.set_ylabel("final tumor volume per regimen")
ax1.boxplot(volumes)
plt.xticks([1,2,3,4], ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.savefig("Analysis Directory/Boxplot.png")
plt.show()
```
